```python
from numpy import array
import numpy as np
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.utils.np_utils import to_categorical
from keras.layers import TimeDistributed
```

#### Generating fake data

```python
num_samples_left = 50
num_sample_right = 50
total_samples = num_samples_left+num_sample_right
time_steps = 10
num_features = 6

X = []
Y = np.zeros((1,total_samples)).astype(str)
for i in range(num_samples_left):
    entry = []
    for j in range(time_steps):
        entry_in_entry = []
        for k in range(num_features):
            entry_in_entry.append(-1)
        entry.append(entry_in_entry)
    X.append(entry)
    Y[0][i]=0
for i in range(num_sample_right):
    entry = []
    for j in range(time_steps):
        entry_in_entry = []
        for k in range(num_features):
            entry_in_entry.append(1)
        entry.append(entry_in_entry)
    X.append(entry)
    Y[0][i+num_samples_left]=1
Y=to_categorical(Y, 2)
```

```python
data = array([[[0.1, 1.0],[0.2, 0.9],[0.3, 0.8],[0.4, 0.7],[0.5, 0.6],
              [0.6, 0.5],[0.7, 0.4],[0.8, 0.3],[0.9, 0.2],[1.0, 0.1]],
             [[-0.1, -1.0],[-0.2, -0.9],[-0.3, -0.8],[-0.4, -0.7],[-0.5, -0.6],
              [-0.6, -0.5],[-0.7, -0.4],[-0.8, -0.3],[-0.9, -0.2],[-1.0, -0.1]]])
```

```python
data = data.reshape(2, 10, 2)
```

```python
y=[0,1]
#y=to_categorical(y, 2)
```

```python
model = Sequential()
model.add(LSTM(32, input_shape=(10, 2)))
#model.add(Dense(1))
model.add(Dense(1))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
print(model.summary())
```

```python
model.fit(data, y, epochs=100, batch_size=2, verbose=0)
```

### One to One Example

```python
length = 5
seq = array([i/float(length) for i in range(length)])
print(seq)
```

```python
X = seq.reshape(5, 1, 1)
y = seq.reshape(5, 1)
```

```python
# prepare sequence
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(len(seq), 1, 1)
y = seq.reshape(len(seq), 1)
# define LSTM configuration
n_neurons = length
n_batch = length
n_epoch = 1000
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(1, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result:
	print('%.1f' % value)
```

### Many-To-One LSTM (without timeDistrubuted)

```python
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length)
# define LSTM configuration
n_neurons = length
n_batch = 1
n_epoch = 500
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1)))
model.add(Dense(length))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:]:
	print('%.1f' % value)
```

### Many-to-Many (with timeDistributed)

```python
length = 5
seq = array([i/float(length) for i in range(length)])
X = seq.reshape(1, length, 1)
y = seq.reshape(1, length, 1)
# define LSTM configuration
n_neurons = length
n_batch = 1
n_epoch = 1000
# create LSTM
model = Sequential()
model.add(LSTM(n_neurons, input_shape=(length, 1), return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(loss='mean_squared_error', optimizer='adam')
print(model.summary())
# train LSTM
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
# evaluate
result = model.predict(X, batch_size=n_batch, verbose=0)
for value in result[0,:,0]:
	print('%.1f' % value)
```
